## Inline Chat Score Method

Inline chat prompt:

```text
Add another method that takes 5 subject scores as subject1, subject2, subject3, subject4, and subject5, then displays the sum, average, highest score, and lowest score.
```

````python
def calculate_score_details(subject1, subject2, subject3, subject4, subject5):
    scores = [subject1, subject2, subject3, subject4, subject5]
    total_score = sum(scores)
    average_score = total_score / len(scores)
    highest_score = max(scores)
    lowest_score = min(scores)

    return {
        "scores": scores,
        "total": total_score,
        "average": average_score,
        "highest": highest_score,
        "lowest": lowest_score,
    }
````
